```python
from langchain_core.prompts import ChatPromptTemplate

system_template = "Translate the following from English into Hindi."

prompt_template = ChatPromptTemplate.from_messages([
    ("system", system_template),
    ("user", "{text}")
])

```

```python
prompt = prompt_template.invoke({"language":"odia","text":"The Women's Cricket World Cup 2025 will showcase the best international women's teams competing for glory.It's set to be an exciting tournament that highlights the growth and power of women's cricket worldwide."})
prompt
```

```python
prompt.to_messages()
```

```python
response = model.invoke(prompt)
print(response.content)
```
